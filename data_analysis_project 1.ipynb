```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score, f1_score
```

```python
iris = load_iris()

X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.5,random_state=42)
```

```python
log_model = LogisticRegression(max_iter=200)

log_model.fit(X_train,y_train)

log_pred = log_model.predict(X_test)

log_acc = accuracy_score(y_test,log_pred)
log_f1 = f1_score(y_test,log_pred,average='weighted')

print("Logistic Regression Accuracy:",log_acc)
print("Logistic Regression F1 Score:",log_f1)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)

knn_model.fit(X_train,y_train)

knn_pred = knn_model.predict(X_test)

knn_acc = accuracy_score(y_test,knn_pred)
knn_f1 = f1_score(y_test,knn_pred,average='weighted')

print("KNN Accuracy:",knn_acc)
print("KNN F1 Score:",knn_f1)
```

```python
dt_model = DecisionTreeClassifier()

dt_model.fit(X_train,y_train)

dt_pred = dt_model.predict(X_test)

dt_acc = accuracy_score(y_test,dt_pred)
dt_f1 = f1_score(y_test,dt_pred,average='weighted')

print("Decision Tree Accuracy:",dt_acc)
print("Decision Tree F1 Score:",dt_f1)
```

```python
results = pd.DataFrame({
    "Algorithm":["Logistic Regression","KNN","Decision Tree"],
    "Accuracy":[log_acc,knn_acc,dt_acc],
    "F1 Score":[log_f1,knn_f1,dt_f1]
})

print(results)
```

```python
import matplotlib.pyplot as plt

results.plot(x="Algorithm", y=["Accuracy","F1 Score"], kind="bar")

plt.title("Algorithm Performance Comparison")
plt.ylabel("Score")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(results["Algorithm"], results["Accuracy"], marker='o')
plt.title("Algorithm Accuracy Comparison")
plt.xlabel("Algorithm")
plt.ylabel("Accuracy")
plt.show()
```
